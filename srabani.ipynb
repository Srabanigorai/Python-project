```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:\\Users\\anubh\\Downloads\\titanic3.csv"
df = pd.read_csv(file_path)

# Calculate survival rate by passenger class
survival_by_class = df.groupby('pclass')['survived'].mean().reset_index()

# Set visual style
sns.set(style="whitegrid")

# Create bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x='pclass', y='survived', data=survival_by_class, palette='Blues_d')

# Customize the plot
plt.title('Survival Rate by Passenger Class', fontsize=16)
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.xticks(ticks=[0, 1, 2], labels=['1st Class', '2nd Class', '3rd Class'])
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "C:\\Users\\anubh\\Downloads\\titanic3.csv"
df = pd.read_csv(file_path)

# Drop rows with missing age or fare values
age_fare_df = df[['age', 'fare']].dropna()

# Calculate the correlation matrix
correlation = age_fare_df.corr()

# Set the visual style
sns.set(style="white")

# Draw a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", square=True, linewidths=0.5)

# Customize the plot
plt.title('Correlation Heatmap: Age vs Fare', fontsize=14)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:\\Users\\anubh\\Downloads\\titanic3.csv"
df = pd.read_csv(file_path)

# Categorize fares into bins
fare_bins = [0, 50, 100, 200, df['fare'].max()]
fare_labels = ['Low (0-50)', 'Medium (50-100)', 'High (100-200)', 'Very High (200+)']
df['fare_category'] = pd.cut(df['fare'], bins=fare_bins, labels=fare_labels, include_lowest=True)

# Count passengers in each fare category
fare_counts = df['fare_category'].value_counts().sort_index()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(fare_counts, labels=fare_counts.index, autopct='%1.1f%%',
        startangle=140, colors=sns.color_palette('pastel'))

# Customize plot
plt.title('Passenger Distribution by Fare Category', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:\\Users\\anubh\\Downloads\\titanic3.csv"
df = pd.read_csv(file_path)

# Get the top 10 oldest passengers
top_10_aged = df[['name', 'age', 'sex', 'pclass', 'survived']].sort_values(by='age', ascending=False).head(10)

# Set the visual style
sns.set(style="whitegrid")

# Create a barplot
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_aged, x='age', y='name', hue='survived', dodge=False, palette='Set2')

# Customize the plot
plt.title('Top 10 Oldest Passengers on the Titanic', fontsize=16)
plt.xlabel('Age')
plt.ylabel('Passenger Name')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:\\Users\\anubh\\Downloads\\titanic3.csv"
df = pd.read_csv(file_path)

# Get the top 10 oldest passengers
top_10_aged = df[['name', 'age', 'sex', 'pclass', 'survived']].sort_values(by='age', ascending=False).head(10)

# Set the visual style
sns.set(style="whitegrid")

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=top_10_aged,
    x='age',
    y='name',
    hue='survived',
    style='sex',
    s=100,
    palette='Set2'
)

# Customize the plot
plt.title('Scatter Plot of Top 10 Oldest Passengers', fontsize=16)
plt.xlabel('Age')
plt.ylabel('Passenger Name')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()

# Show the plot
plt.show()
```
